```python
# library to handle requests
import requests

#importing beautifulsoup
from bs4 import BeautifulSoup

# library for data analsysis
import pandas as pd

# library to handle data in a vectorized manner
import numpy as np

 # convert an address into latitude and longitude values
from geopy.geocoders import Nominatim

# libraries for displaying images
from IPython.display import Image 
from IPython.core.display import HTML 

#!conda install -c conda-forge folium=0.5.0 --yes
import folium # plotting library

# import k-means from clustering stage
from sklearn.cluster import KMeans

# importing package to assign colors on maps
import matplotlib.cm as cm
import matplotlib.colors as colors
```

Scraping Toronto Neighborhoods Data

```python
source = requests.get('https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=945633050.').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

```python
# Dropping the rows where Borough is 'Not assigned'
df1 = df[df.Borough != 'Not assigned']

# Combining the neighbourhoods with same Postalcode
df2 = df1.groupby(['Postcode','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned',df2['Borough'], df2['Neighbourhood'])

df2
```

Determining Shape of the Dataframe containing Toronto Neighborhoods data

```python
df2.shape
```

Importing geospatial data for toronto postal codes from the weblink

```python
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

Joining Lat Long information to Tornoto Neighborhoods data using Postal Code as key

```python
df3 = pd.merge(df2,lat_lon,left_on='Postcode', right_on='Postal Code')
del df3['Postal Code']
df3.head()
```

Determining neighborhoods in Toronto

```python
df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
```

Determining Lat & Long for Toronto

```python
address = 'Toronto, CA'

geolocator = Nominatim(user_agent="tornoto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

Visualizing Neighborhoods of Tornoto

```python
toronto_visual = folium.Map(location=[latitude,longitude],zoom_start=11)
#toronto_visual = folium.Map(location=[43.651070,-79.347015],zoom_start=12)

for lat,lng,borough,neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='purple',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(toronto_visual)
toronto_visual
```

Clustering Neighborhoods of Toronto

```python
k=5
toronto_cluster = df4.drop(['Postcode','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_cluster)
kmeans.labels_
df4.insert(0, 'Cluster Labels', kmeans.labels_)
```

Checking to determine if cluster has been added to dataframe with Nighborhood Info

```python
df4
```

```python
# creating cluster map
map_clusters = folium.Map(location=[latitude,longitude],zoom_start=11)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighbourhood'], df4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
